```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../DATASET/dataset_expandido.csv")
df
```

```python
#Criar uma coluna nova de JobSat para Classification
bins   = [-1, 5, 7, 10]
labels = ['Baixo', 'Medio', 'Alto']

df['JobSat_Class'] = pd.cut(df['JobSat'], bins=bins, labels=labels)

# Verificar resultado
print(df['JobSat_Class'].value_counts())
print(df[['JobSat', 'JobSat_Class']].head(10))

# Guardar dataset
df.to_csv('../DATASET/dataset_expandido.csv', index=False)
```
