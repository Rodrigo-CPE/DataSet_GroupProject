```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
import plotly.express as px
from plotly.subplots import make_subplots

# Dados extraídos do survey (simulando o DataFrame)
# Na prática, carregarias com: df = pd.read_csv('survey.csv')

# ============================================
# 1. SENTIMENTO SOBRE IA (AISent)
# ============================================

ai_sentiment_data = {
    'Sentimento': ['Very favorable', 'Favorable', 'Neutral', 'Unfavorable', 'Very unfavorable'],
    'Percentagem': [45, 30, 15, 7, 3]  # Valores exemplo - substitui pelos teus dados reais
}

fig_sentiment = go.Figure(data=[
    go.Bar(
        x=ai_sentiment_data['Sentimento'],
        y=ai_sentiment_data['Percentagem'],
        marker_color=['#2ecc71', '#27ae60', '#95a5a6', '#e74c3c', '#c0392b'],
        text=ai_sentiment_data['Percentagem'],
        texttemplate='%{text}%',
        textposition='outside'
    )
])



# ============================================
# 2. FERRAMENTAS DE IA EM USO (AIToolCurrently Using)
# ============================================

ai_tools = {
    'Ferramenta': ['ChatGPT', 'GitHub Copilot', 'Bing AI', 'Claude', 'Bard', 'Tabnine', 'AWS CodeWhisperer'],
    'Utilizadores': [78, 45, 32, 28, 25, 15, 12]  # Valores exemplo
}

fig_tools = go.Figure(data=[
    go.Bar(
        y=ai_tools['Ferramenta'],
        x=ai_tools['Utilizadores'],
        orientation='h',
        marker_color=px.colors.sequential.Viridis,
        text=ai_tools['Utilizadores'],
        texttemplate='%{text}%',
        textposition='outside'
    )
])



# ============================================
# 3. BENEFÍCIOS PERCEBIDOS DA IA (AIBen)
# ============================================

ai_benefits = {
    'Benefício': ['Aumentar produtividade', 'Maior eficiência', 'Acelerar aprendizagem', 
                  'Melhorar precisão', 'Melhorar colaboração'],
    'Respostas': [85, 72, 65, 48, 35]  # Valores exemplo
}

fig_benefits = go.Figure(data=[
    go.Pie(
        labels=ai_benefits['Benefício'],
        values=ai_benefits['Respostas'],
        hole=0.4,
        marker_colors=px.colors.sequential.Teal
    )
])



# ============================================
# 4. CASOS DE USO DE IA (AIToolCurrently Using - detalhado)
# ============================================

use_cases = {
    'Caso de Uso': ['Writing code', 'Debugging', 'Search for answers', 
                   'Learning about codebase', 'Documenting code', 
                   'Project planning', 'Testing code', 'Deployment'],
    'A usar': [75, 68, 82, 55, 45, 38, 25, 18],
    'Interessado': [15, 20, 10, 30, 35, 40, 45, 50]
}

fig_usecases = go.Figure()

fig_usecases.add_trace(go.Bar(
    name='A usar atualmente',
    x=use_cases['Caso de Uso'],
    y=use_cases['A usar'],
    marker_color='#3498db'
))

fig_usecases.add_trace(go.Bar(
    name='Interessado em usar',
    x=use_cases['Caso de Uso'],
    y=use_cases['Interessado'],
    marker_color='#9b59b6'
))



# ============================================
# 5. NÍVEL DE CONFIANÇA NA IA (AIAcc)
# ============================================

ai_trust = {
    'Nível': ['Highly trust', 'Somewhat trust', 'Neither trust nor distrust', 
              'Somewhat distrust', 'Highly distrust'],
    'Percentagem': [8, 52, 25, 12, 3]  # Valores exemplo
}

fig_trust = go.Figure(data=[
    go.Funnel(
        y=ai_trust['Nível'],
        x=ai_trust['Percentagem'],
        textinfo='value+percent total',
        marker_color=['#2ecc71', '#27ae60', '#f39c12', '#e67e22', '#e74c3c']
    )
])



# ============================================
# 6. DESAFIOS COM IA (AIChallenges)
# ============================================

challenges = {
    'Desafio': ['Não confio no output', 'Falta de contexto', 'Problemas de privacidade',
                'Desinformação', 'Atribuição incorreta', 'Custo das ferramentas'],
    'Menções': [65, 55, 48, 42, 38, 25]  # Valores exemplo
}

fig_challenges = go.Figure(data=[
    go.Scatter(
        x=challenges['Menções'],
        y=challenges['Desafio'],
        mode='markers',
        marker=dict(
            size=challenges['Menções'],
            color=challenges['Menções'],
            colorscale='Reds',
            showscale=True
        )
    )
])



# ============================================
# 7. FERRAMENTAS DE DESENVOLVIMENTO (IDE/Editors)
# ============================================

dev_tools = {
    'Ferramenta': ['VS Code', 'Visual Studio', 'IntelliJ IDEA', 'PyCharm', 
                   'WebStorm', 'Vim/Neovim', 'Sublime Text'],
    'Utilização': [74, 32, 28, 22, 15, 18, 12]  # Valores exemplo
}

fig_devtools = go.Figure(data=[
    go.Bar(
        x=dev_tools['Ferramenta'],
        y=dev_tools['Utilização'],
        marker_color=px.colors.sequential.Blues_r,
        text=dev_tools['Utilização'],
        texttemplate='%{text}%',
        textposition='outside'
    )
])



# ============================================
# 8. FUTURO DA IA NO DESENVOLVIMENTO (AINext)
# ============================================

ai_future = {
    'Expectativa': ['Muito mais integrada', 'Mais integrada', 'Sem mudança', 
                    'Menos integrada', 'Muito menos integrada'],
    'Percentagem': [35, 45, 15, 4, 1]  # Valores exemplo
}

fig_future = go.Figure(data=[
    go.Pie(
        labels=ai_future['Expectativa'],
        values=ai_future['Percentagem'],
        marker_colors=['#1abc9c', '#16a085', '#95a5a6', '#e74c3c', '#c0392b'],
        textinfo='label+percent',
        pull=[0.1, 0, 0, 0, 0]
    )
])



# ============================================
# GUARDAR PARA DASHBOARD
# ============================================

# Para Dash/Streamlit, podes usar diretamente os objetos fig_*
# Para HTML standalone:
fig_sentiment.write_html('ai_sentiment.html')
fig_tools.write_html('ai_tools.html')
fig_benefits.write_html('ai_benefits.html')
fig_usecases.write_html('ai_usecases.html')
fig_trust.write_html('ai_trust.html')
fig_challenges.write_html('ai_challenges.html')

# Mostrar um gráfico (em Jupyter)
fig_sentiment.show()
```
